# Cleaning/handling missing values 

```python
df["Cabin"].unique()
```

```python
colums_drop = ['Name', 'Ticket','Fare']
new_df = df.drop(colums_drop, axis=1)
```

```python
new_df
```

```python
mode_cabin = new_df['Cabin'].mode()[0]
new_df['Cabin'].fillna(mode_cabin, inplace=True)
```

```python
#new_df.head(30)
```

```python
new_df.isna().sum()
```

```python
age_mean = new_df["Age"].mean() # Replcaing nan age values by calculating mean 
new_df["Age"].fillna(age_mean,inplace=True)
#new_df.head(30)
```

```python
most_frequent_value = new_df['Embarked'].mode()[0] # Replacing nan Embarked value by taking mode(value that occurs most of the time)
new_df['Embarked'].fillna(most_frequent_value, inplace=True)
```

```python
# new_df.isna().sum()
```

```python
new_df.head()
```

```python
new_df["Sex"].replace({"male":"M","female":"F"},inplace=True)
new_df.head()
```

# ENCODING DATA

```python
from sklearn.preprocessing import OneHotEncoder # Encoding using one-hot sklearn function 
data = new_df["Embarked"]
df = pd.DataFrame(data)
encoder = OneHotEncoder(sparse=False)
encoded_columns = encoder.fit_transform(df[['Embarked']])
encoded_df = pd.DataFrame(encoded_columns, columns=encoder.get_feature_names_out(['Embarked']))
new_df = pd.concat([new_df, encoded_df], axis=1).drop(columns=['Embarked'])
new_df.head(5)
```

```python
new_df.drop("Cabin",axis=1,inplace=True)
new_df.head()
```

```python
# new_df = pd.get_dummies(new_df, columns=['Sex']) # Encoding using pandas
# new_df.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder # We use label encoding so that dimensionality does not increase as only 2 values(Male & Female are present in dataset)
label_encoder = LabelEncoder()
new_df['Sex'] = label_encoder.fit_transform(new_df['Sex'])
```
